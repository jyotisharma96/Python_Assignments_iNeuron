1.Implement a class iterator to flatten a nested list of lists of integers. Each
list element is either an integer or a list. There can be many levels of nested
lists in lists.

The class initializes with a nested list. It also has two methods:

1. next() returns an integer in the order of appearance.

2. hasNext() returns True / False regarding if all integers have been
retrieved or not.

Write the Class implementation for three required methods.

Examples:
    
ni, actual = NestedIterator([[1, 1], 2, [1, 1]]), []

while ni.hasNext():

actual.append(ni.next())

actual ➞ [1, 1, 2, 1, 1]

ni, actual = NestedIterator([1, [4, [6]]]), []

while ni.hasNext():

actual.append(ni.next())

actual ➞ [1, 4, 6]

ni, actual = NestedIterator([[[]], []]), []

while ni.hasNext():

actual.append(ni.next())

actual ➞ []

```python
class NestedIterator:
    def __init__(self, nestedList):
        self.flatten_list = self.flatten(nestedList)
        self.index = 0

    def flatten(self, nestedList):
        result = []
        for item in nestedList:
            if isinstance(item, int):
                result.append(item)
            else:
                result.extend(self.flatten(item))
        return result

    def next(self):
        if self.hasNext():
            value = self.flatten_list[self.index]
            self.index += 1
            return value

    def hasNext(self):
        return self.index < len(self.flatten_list)

ni = NestedIterator([[1, 1], 2, [1, 1]])
actual = []
while ni.hasNext():
    actual.append(ni.next())
print(actual)

ni = NestedIterator([1, [4, [6]]])
actual = []
while ni.hasNext():
    actual.append(ni.next())
print(actual)

ni = NestedIterator([[[]], []])
actual = []
while ni.hasNext():
    actual.append(ni.next())
print(actual)
```

2.Implement the class Shape that receives perimeter and density function
into __init__ method. The list of consecutive corners defines shape of a 2-
dimensional object. The density function defines the mass distribution inside
the shape. To compute mass in a certain point m(x, y) = small_square *
density(x, y). The __init__ method calls other internal methods that compute
three characteristics of the shape:

- area

- total mass

- center of mass (xc, yc)

The computational grid has distance between two neighboring points as 2 *
delta, the distance between a grid point and the perimeter wall is delta.

Examples:

sh_ex1 = Shape([(1, 1), (3, 1), (3, 2), (1, 2)], lambda x, y: 100 + 100 * x)

sh_ex1.area ➞ 2.0

sh_ex1.mass ➞ 600.0

sh_ex1.mass_center ➞ (2.1, 1.5)

The example can be verified via analytical integration. Other shapes in Tests
are slightly more complicated and require numerical integration as illustrated
here:


```python
class Shape:
    def __init__(self, corners, density):
        self.corners = corners
        self.density = density
        self.area = self.compute_area()
        self.mass = self.compute_mass()
        self.mass_center = self.compute_mass_center()

    def compute_area(self):
        x_coords = [corner[0] for corner in self.corners]
        y_coords = [corner[1] for corner in self.corners]
        return abs(max(x_coords) - min(x_coords)) * abs(max(y_coords) - min(y_coords))

    def compute_mass(self):
        delta = 0.001  # small delta for discretization
        total_mass = 0.0
        for x in self.range_x():
            for y in self.range_y():
                if self.is_inside(x, y):
                    small_square = (2 * delta) ** 2
                    total_mass += small_square * self.density(x, y)
        return total_mass

    def compute_mass_center(self):
        delta = 0.001  # small delta for discretization
        total_mass_x = 0.0
        total_mass_y = 0.0
        total_mass = 0.0
        for x in self.range_x():
            for y in self.range_y():
                if self.is_inside(x, y):
                    small_square = (2 * delta) ** 2
                    mass = small_square * self.density(x, y)
                    total_mass += mass
                    total_mass_x += x * mass
                    total_mass_y += y * mass
        xc = total_mass_x / total_mass
        yc = total_mass_y / total_mass
        return xc, yc

    def range_x(self):
        return range(min(corner[0] for corner in self.corners) - 1, max(corner[0] for corner in self.corners) + 2)

    def range_y(self):
        return range(min(corner[1] for corner in self.corners) - 1, max(corner[1] for corner in self.corners) + 2)

    def is_inside(self, x, y):
        crossings = 0
        for i in range(len(self.corners)):
            j = (i + 1) % len(self.corners)
            if ((self.corners[i][1] > y) != (self.corners[j][1] > y)) and (x < (self.corners[j][0] - self.corners[i][0]) * (y - self.corners[i][1]) / (self.corners[j][1] - self.corners[i][1]) + self.corners[i][0]):
                crossings += 1
        return crossings % 2 == 1

sh_ex1 = Shape([(1, 1), (3, 1), (3, 2), (1, 2)], lambda x, y: 100 + 100 * x)
print(sh_ex1.area) 
print(sh_ex1.mass) 
print(sh_ex1.mass_center)
```

3.Given a 3x3 matrix of a completed tic-tac-toe game, create a function that
returns whether the game is a win for &quot;X&quot;, &quot;O&quot;, or a &quot;Draw&quot;, where &quot;X&quot; and &quot;O&quot;
represent themselves on the matrix, and &quot;E&quot; represents an empty spot.

Examples:
    
tic_tac_toe([
[&quot;X&quot;, &quot;O&quot;, &quot;X&quot;],
[&quot;O&quot;, &quot;X&quot;, &quot;O&quot;],
[&quot;O&quot;, &quot;X&quot;, &quot;X&quot;]
]) ➞ &quot;X&quot;

tic_tac_toe([
[&quot;O&quot;, &quot;O&quot;, &quot;O&quot;],
[&quot;O&quot;, &quot;X&quot;, &quot;X&quot;],
[&quot;E&quot;, &quot;X&quot;, &quot;X&quot;]
]) ➞ &quot;O&quot;

tic_tac_toe([
[&quot;X&quot;, &quot;X&quot;, &quot;O&quot;],
[&quot;O&quot;, &quot;O&quot;, &quot;X&quot;],
[&quot;X&quot;, &quot;X&quot;, &quot;O&quot;]
]) ➞ &quot;Draw&quot;

```python
def tic_tac_toe(matrix):
    # Check rows
    for row in matrix:
        if row[0] == row[1] == row[2] != 'E':
            return row[0]

    # Check columns
    for col in range(3):
        if matrix[0][col] == matrix[1][col] == matrix[2][col] != 'E':
            return matrix[0][col]

    # Check diagonals
    if matrix[0][0] == matrix[1][1] == matrix[2][2] != 'E':
        return matrix[0][0]
    if matrix[0][2] == matrix[1][1] == matrix[2][0] != 'E':
        return matrix[0][2]

    # If no winner and no empty spots, it's a draw
    if all('E' not in row for row in matrix):
        return "Draw"

    # If no winner and there are still empty spots, the game is not finished
    return "Game not finished"

matrix1 = [
    ["X", "O", "X"],
    ["O", "X", "O"],
    ["O", "X", "X"]
]
print(tic_tac_toe(matrix1))

matrix2 = [
    ["O", "O", "O"],
    ["O", "X", "X"],
    ["E", "X", "X"]
]
print(tic_tac_toe(matrix2))

matrix3 = [
    ["X", "X", "O"],
    ["O", "O", "X"],
    ["X", "X", "O"]
]
print(tic_tac_toe(matrix3))
```

4.Your computer might have been infected by a virus! Create a function that
finds the viruses in files and removes them from your computer.

Examples:

remove_virus(&quot;PC Files: spotifysetup.exe, virus.exe, dog.jpg&quot;) ➞ &quot;PC Files:

spotifysetup.exe, dog.jpg&quot;

remove_virus(&quot;PC Files: antivirus.exe, cat.pdf, lethalmalware.exe,

dangerousvirus.exe &quot;) ➞ &quot;PC Files: antivirus.exe, cat.pdf&quot;

remove_virus(&quot;PC Files: notvirus.exe, funnycat.gif&quot;) ➞ &quot;PC Files:

notvirus.exe, funnycat.gif&quot;)

```python
def remove_virus(files):
    cleaned_files = []
    file_list = files.split(': ')[1].split(', ')
    
    for file in file_list:
        if not file.endswith('.exe'):
            cleaned_files.append(file)
    
    return 'PC Files: ' + ', '.join(cleaned_files)

files1 = "PC Files: spotifysetup.exe, virus.exe, dog.jpg"
print(remove_virus(files1))

files2 = "PC Files: antivirus.exe, cat.pdf, lethalmalware.exe, dangerousvirus.exe"
print(remove_virus(files2))

files3 = "PC Files: notvirus.exe, funnycat.gif"
print(remove_virus(files3))
```
