1)What is the difference between enclosing a list comprehension in square brackets and
parentheses?

Answer- When using a list comprehension in Python, you can enclose it either in square brackets or parentheses. The choice between these two determines the type of object that is created.

Square brackets: When you enclose a list comprehension in square brackets, it creates a new list. The resulting object is a list containing the elements generated by the comprehension.


```python
my_list = [x for x in range(5)]
print(my_list)  # Output: [0, 1, 2, 3, 4]
```

In this case, my_list is a list object containing the numbers 0 to 4 generated by the list comprehension.

Parentheses: When you enclose a list comprehension in parentheses, it creates a generator object. A generator is an iterable that produces values on-the-fly, as they are needed. It doesn't store all the values in memory at once, which can be useful when dealing with large datasets.

```python
my_generator = (x for x in range(5))
print(my_generator)  # Output: <generator object <genexpr> at 0xXXXXXXX>

```

In this case, my_generator is a generator object that yields the numbers 0 to 4 as you iterate over it. You can use it in a for loop or with other iterable operations. The values are generated on-demand and not stored in memory all at once.

2)What is the relationship between generators and iterators?

Answer- Generators and iterators are closely related concepts in Python, as generators are a specific type of iterator. Understanding iterators is important to grasp the relationship between the two.

Iterators: In Python, an iterator is an object that implements the iterator protocol, which consists of two methods: __iter__() and __next__(). The __iter__() method returns the iterator object itself, and the __next__() method returns the next element from the iterator. When there are no more elements to return, the __next__() method raises the StopIteration exception.
Iterators allow you to iterate over a sequence of elements one at a time, without storing the entire sequence in memory. They provide a memory-efficient and lazy way to access elements sequentially. Iterators are used implicitly in for loops, and you can create your own iterators by implementing the iterator protocol.

```python
# iterator
iter_list = iter(['Apple', 'Orange', 'Banana'])
print(next(iter_list))
print(next(iter_list))
print(next(iter_list))
```

Generators: Generators are a special type of iterator. They are defined using a special syntax in Python, which is the use of a function with the yield statement. When a generator function is called, it returns a generator object, which is an iterator. The generator object can be iterated over using the same iterator protocol mentioned earlier.

Generators differ from regular functions in that they do not use the return statement to return a value. Instead, they use yield to produce a sequence of values. When a yield statement is encountered, the generator function pauses its execution, saves its state, and returns the yielded value. The next time the generator is called, it resumes from where it left off and continues executing until the next yield statement.

```python
# Example of Generator
def cube_numbers(in_num):
    for ele in range(in_num+1):
        yield ele**3

out_num = cube_numbers(4)
print(next(out_num))
print(next(out_num))
print(next(out_num))
print(next(out_num))
print(next(out_num))
```

4)What is the purpose of a yield statement?

Answer - The yield statement in Python serves the purpose of creating generator functions. When used within a function, yield allows the function to generate a series of values one at a time, rather than computing and returning all the values at once.

5)What is the relationship between map calls and list comprehensions? Make a comparison and
contrast between the two.

Answer- 

1. Map calls: Suppose we have a function and we want to compute this function for different values in a single line of code . This is where map() function plays its role. map() function returns a map object(which is an iterator) of the results after applying the given function to each item of a given iterable (list, tuple etc.) 

        If we already have a function defined, it is often good to use map For example, map(sum, myLists) is more neat than [sum(x) for x in myLists]. You gain the elegance of not having to make up a dummy variable (e.g. sum(x) for x... or sum(_) for _... or sum(readableName) for readableName...) which you have to type twice, just to iterate.

2. List Comprehension is a substitute for the lambda function, map(), filter() and reduce().



The contrast between map calls and list comprehensions :

    1. List comprehension is more concise and easier to read as compared to map
    2. List comprehension allows filtering. In map, we have no such facility
    For example, to print all even numbers in range of 100, we can write [n for n in range(100) if n%2 == 0]. There is 
    no alternate for it in map
    3. List comprehension are used when a list of results is required, where as map only returns a map object and does 
    not return any list.
    4. List comprehension is faster than map when we need to evaluate expressions that are too long or complicated 
    to express
    5. Map is faster in case of calling an already defined function (as no lambda is required)
